```python
########################################
#
#
# DONE: Pinterest auto-post (2021.02.10)
# login policy: a few minutes (10?)
#
########################################

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup
import urllib.request
import os

driver = None

def get_driver():
    # xattr -d com.apple.quarantine chromedriver
    CHROME_EXE_PATH = "./chromedriver"
    
    return webdriver.Chrome(executable_path=CHROME_EXE_PATH)

# return False when driver is None else False
def login(driver=None, email="nogne@naver.com", passwd="nsw1121*"):
    if driver is None: return False
    
    URL_LOGIN = 'https://www.pinterest.co.kr/'
    INTERVAL_LOAD_LOGIN_PAGE = 3
    INTERVAL_SHORT = 1
    
    driver.get(URL_LOGIN)
    
    print("Wait for loading a login page...")
    time.sleep(INTERVAL_LOAD_LOGIN_PAGE) 
    
    # click login
    print ("Open login popup")
    driver.find_element_by_xpath('/html/body/div[1]/div[1]/div/div/div/div[1]/div[1]/div[2]/div[2]/button').click()
    time.sleep(INTERVAL_SHORT)
    
    email_box = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div/div/div/div[1]/div[2]/div[2]/div/div/div/div/div/div/div/div[4]/form/div[1]/fieldset/span/input')
    email_box.send_keys(email)
    passwd_box = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div/div/div/div[1]/div[2]/div[2]/div/div/div/div/div/div/div/div[4]/form/div[2]/fieldset/span/input')
    passwd_box.send_keys(passwd)    
    passwd_box.submit()

    print("Login info has been submitted.")
    return True

def create_new_post(driver, img="/Users/macbook/Desktop/dev/python/sns/upload_imgs/snake.jpg"):
    
    INTERVAL_LONG = 30
    INTERVAL_SHORT = 5
    time.sleep(INTERVAL_LONG)
    
    print ("Try clicking new post button...")
    
    # click + button
    driver.find_element_by_xpath("/html/body/div[1]/div[1]/div[3]/div/div[1]/div[1]/div/div/div/button")\
        .click()
    
    time.sleep(INTERVAL_LONG)
    
    # cancel button
    driver.find_element_by_xpath("/html/body/div[3]/div/div/div/div/div[2]/div/div/div/div[3]/div/button")\
        .click()
    
    time.sleep(INTERVAL_LONG)
    
    # click + button
    driver.find_element_by_xpath("/html/body/div[1]/div[1]/div[3]/div/div[1]/div[1]/div/div/div/button")\
        .click()
    
    time.sleep(INTERVAL_SHORT)
    
    # create a new pin
    driver.find_element_by_xpath("/html/body/div[1]/div[1]/div[3]/div/div[1]/div[1]/div[2]/div/div/div/div/div/div[2]/div/div/div/div/div")\
        .click()
    
    time.sleep(INTERVAL_SHORT)
    
    # upload image
    driver.find_element_by_xpath("/html/body/div[1]/div[1]/div[2]/div/div/div/div[2]/div[1]/div/div/div/div/div/div/div[2]/div/div[1]/div/div/div[1]/div/div/div/input")\
        .send_keys(img)
    
    time.sleep(INTERVAL_LONG)
    
    driver.find_element_by_xpath("/html/body/div[1]/div[1]/div[2]/div/div/div/div[2]/div[1]/div/div/div/div/div/div/div[2]/div/div[2]/div/div[1]/div[1]/div/div/div[1]/textarea")\
        .send_keys("Title test")

    time.sleep(INTERVAL_LONG)
    
    driver.find_element_by_xpath("/html/body/div[1]/div[1]/div[2]/div/div/div/div[2]/div[1]/div/div/div/div/div/div/div[1]/div/div[2]/div/div/div/button[2]")\
        .click()

    time.sleep(INTERVAL_LONG)
    
    # do more
    print ("Finish clicking new post button!")


```

```python
def main():
    driver = get_driver()
    is_success = login(driver)
    if is_success is False: return

    create_new_post(driver)


if __name__ == "__main__":
    main()
```
